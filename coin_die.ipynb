```python
import numpy as np
import matplotlib.pyplot as plt

n = 1000
U = np.random.rand(n)
toss = (U < 0.5).astype(int)  # 1 para cara, 0 para coroa
a = np.zeros(n + 1)
avg = np.zeros(n)

for i in range(1, n + 1):
    a[i] = a[i - 1] + toss[i - 1]
    avg[i - 1] = a[i] / i

plt.plot(avg)
plt.grid(True)
plt.xlabel('Número de Lançamentos')
plt.ylabel('Proporção de Caras')
plt.title('Lei dos Grandes Números (Simulação de Lançamento de Moeda)')
plt.show()
```

**Explicações:**

---
*clc*

Em MATLAB limpa a janela de comandos.

Em Python, essa funcionalidade não é diretamente necessária no código em si, pois a saída do gráfico será em uma janela separada. Se você estiver executando em um ambiente interativo (como um notebook Jupyter ou um terminal IPython), você pode usar comandos específicos do ambiente para limpar a saída anterior.

---

*rand(1, n)*

Em MATLAB, isso cria um vetor linha de n números aleatórios uniformemente distribuídos entre 0 e 1.

Em Python, usamos np.random.rand(n) para criar um array NumPy unidimensional (equivalente a um vetor linha) com as mesmas propriedades.

---

*(U < 0.5)'*

Em MATLAB, U < 0.5 cria um vetor booleano (1 para verdadeiro, 0 para falso) onde os elementos de U são menores que 0.5. O apóstrofo ' realiza a transposição, convertendo o vetor linha em um vetor coluna.

Em Python, U < 0.5 também cria um array booleano. Usamos .astype(int) para converter os valores booleanos para inteiros (True -> 1, False -> 0), representando cara (1) e coroa (0). A transposição não é estritamente necessária aqui, pois o loop subsequente funciona bem com um array unidimensional.

---

*zeros(n + 1) e zeros(n)*

Em MATLAB, zeros(m, n) cria uma matriz de zeros de tamanho m x n. Se apenas um argumento é fornecido, cria um vetor linha de zeros.

Em Python, np.zeros(shape) cria um array NumPy preenchido com zeros. Usamos np.zeros(n + 1) e np.zeros(n) para criar arrays unidimensionais de zeros com os tamanhos correspondentes.


---


*Loop for i = 2 : n + 1*

Em MATLAB, a indexação começa em 1.

Em Python, a indexação começa em 0. Portanto,
o loop for i = 2 : n + 1 em MATLAB é equivalente a for i in range(1, n + 1) em Python.

Note que estamos acessando toss(i - 1) em MATLAB, que corresponde a toss[i - 1] em Python devido à indexação baseada em zero.


---


*a(i) = a(i - 1) + toss(i - 1)*

A lógica permanece a mesma: acumula o número de caras.


---


*avg(i - 1) = a(i)/(i - 1)*

Calcula a proporção de caras até o lançamento atual. Novamente, ajustamos o índice para Python.


---


plot(avg)

Em MATLAB, plot(x) plota os valores do vetor x em relação aos seus índices.

Em Python, plt.plot(avg) faz o mesmo, usando os índices do array avg como o eixo x e os valores de avg como o eixo y.


---


grid

Em MATLAB, grid ativa a grade no gráfico.

Em Python, usamos plt.grid(True) para obter o mesmo efeito.
